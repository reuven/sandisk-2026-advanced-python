# Exercise: Bookstore

1. Use `namedtuple` to create a `Book` class. Each instance of `Book` should have `title`, `author`, and `price`.
2. Create 3 instances of `Book`
3. Allow a user to ask if the store carries the book.
    - If we get an empty string, then we'll stop asking, and print the total cost.
    - If we get the name of a book in stock, then print its info and add its price to `total`
    - If the book isn't in stock, scold the user


```python
from collections import namedtuple

Book = namedtuple('Book', 'title author price')
b1 = Book('title1', 'author1', 50)
b2 = Book('title2', 'author1', 70)
b3 = Book('title3', 'author2', 60)

inventory = [b1, b2, b3]
total = 0

while True:
    text = input('Enter title: ').strip()

    if not text:    # Did I get an empty string?
        break

    found_book = False
    
    for one_book in inventory:
        if one_book.title == text:
            print(f'Found {one_book.title} by {one_book.author}')
            total += one_book.price
            print(f'\tAdded {one_book.price} to total, which is {total}')
            found_book = True

    if not found_book:
        print(f'No such book {text}')

print(f'Total = {total}')
```

# Exercise: Rainfall

1. Define an empty dict, `rainfall`. The keys will be city names (strings), and the values will be lists of integers.
2. Ask the user, repeatedly, what city they're reporting
    - If we get an empty string, exit from the loop
3. Ask the user for yesterday's rainfall
    - If we've seen this city in the past, add to existing list
    - If we've never seen it, then add a new list with this amount
4. In the end:
    - Print each city name
    - Print total rain and mean rain (`sum` and `len` will help) for each city

```python
rainfall = {}

while city_name := input('City: ').strip():
    mm_rain = input('Rain: ').strip()
    mm_rain = int(mm_rain)

    rainfall.setdefault(city_name, [])
    rainfall[city_name].append(mm_rain)

for key, value in rainfall.items():
    print(f'{key}: total is {sum(value)}, mean is {sum(value)/len(value)}')
```

# Exercise: Travel

1. The end result will be: A dict in which keys are countries (strings) and values are lists of cities (strings).
2. In each iteration, the user will enter `city,country`.
3. Split apart city/country
4. Use `defaultdict` to make it easier.
5. In the end, print each country, and then each city in each country.

```python
all_places = defaultdict(list)

while s := input('Enter city, country: ').strip():
    if s.count(',') == 0:
        print(f'Use "city, country" format')
        continue    

    city, country = s.split(',', maxsplit=1)  # break max once
    all_places[country.strip()].append(city.strip())

for country, all_cities in all_places.items():
    print(country)
    for one_city in all_cities:
        print(f'\t{one_city}')
```

# Parameter types

1. Mandatory argument, either positional or keyword
2. Optional argument, either positional or keyword
3. `*args`, all remaining positional arguments
4. Mandatory keyword-only arguments
5. Optional keyword-only arguments, with defaults

# `**kwargs` -- all unassigned keyword arguments

- This is a dict
- All names of unassigned keyword arguments become strings
- The values are assigned as they were

```python

def myfunc(x, y=2, **kwargs):
    return f'{x=}, {y=}, {kwargs=}'
```

```python
myfunc(10)
```

```python
myfunc(10, 20)
```

```python
myfunc(10, 20, 30)
```

```python
myfunc(10, 20, a=30, b=40)
```

```python
myfunc(10, p=22, q=33, y=44)
```

# Parameter types

1. Mandatory argument, either positional or keyword
2. Optional argument, either positional or keyword
3. `*args`, all remaining positional arguments
4. Mandatory keyword-only arguments
5. Optional keyword-only arguments, with defaults
6. `**kwargs`, dict of unassigned keyword arguments

```python
# what if I want keyword-only arguments, but not *args?

def sub(first, second):
    return first - second
```

```python
sub(10, 3)
```

```python
sub(second=3, first=10)
```

```python
# The * means: after here, only keyword arguments

def sub(*, first, second):
    return first - second
```

```python
sub(10, 3)
```

```python
sub(first=10, second=3)
```

```python
def write_pairs(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}:{value}')
```

```python
write_pairs(a=10, b=20, c=30)
```

```python
# now, change it to write to a file!
def write_pairs(filename, **kwargs):
    with open(filename, 'w') as f:
        for key, value in kwargs.items():
            f.write(f'{key}:{value}\n')
```

```python
write_pairs('config.txt', a=10, b=20, c=30)
```

```python
!cat config.txt
```

```python
write_pairs('config.txt', a=10, b=20, c=30, filename='/etc/passwd')
```
