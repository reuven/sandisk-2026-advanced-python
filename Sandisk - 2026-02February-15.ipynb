# Scoping

```python
x = 100

for i in range(10):
    x = i ** 2

print(x)    
```

```python
i
```

```python
x = 100

def myfunc():
    print(f'{locals()=}')
    print(f'In myfunc, x = {x}')  # is x local? NO. is x global? YES, 100

print(f'Before, x = {x}')       # is there a  global x? Yes, 100
myfunc()
print(f'After, x = {x}')      # is there a  global x? Yes, 100
```

```python
'x' in globals()
```

```python
myfunc.__code__.co_varnames
```
