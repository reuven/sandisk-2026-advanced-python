````python
def a():
    return 'a'

def b():
    return 'b'

ops = {'a':a,
       'b':b}
    
while True:
    choice = input('Choose a or b: ').strip()

    if not choice:
        break

    if choice in ops:
        print(ops[choice]())
    else:
        print(f'Illegal choice!')
        
````

# Exercise: Calculator

1. Let the user enter a math expression with `+` or `-`.
2. Use a dispatch table to call the appropriate function for the operator.
3. If the operator doesn't match what we know, then scold the user.

Example:

    Enter expression: 2 + 10
    12



````python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

ops = {'+':add,
       '-':sub}

while True:
    s = input('Enter expression: ').strip()

    if not s:
        break

    first, op, second = s.split()
    first = int(first)
    second = int(second)

    if op in ops:
        result = ops[op](first, second)
    else:
        result = f'Illegal operator {op}'

    print(f'{first} {op} {second} = {result}')   

    
````

````python
import operator

ops = {'+':operator.add,
       '-':operator.sub}

while True:
    s = input('Enter expression: ').strip()

    if not s:
        break

    first, op, second = s.split()
    first = int(first)
    second = int(second)

    if op in ops:
        result = ops[op](first, second)
    else:
        result = f'Illegal operator {op}'

    print(f'{first} {op} {second} = {result}')   

    
````

# Exercise: Shoe dicts

1. Read the shoe info from `shoe-info.txt`, using a list comprehension
2. The result should be a dict.
3. Every dict should have `brand`, `color`, and `size` as keys.

```python

    'color':'orange',
    'size':'43'},

...

````python
def line_to_dict(one_line):
    fields = one_line.strip().split('\t')
    return {'brand': fields[0],
           'color': fields[1],
           'size': fields[2]
           }

[line_to_dict(one_line)
for one_line in open('shoe-data.txt')]
````
